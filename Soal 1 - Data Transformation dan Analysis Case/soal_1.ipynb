```python
# import library/packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Turn .csv file into dataframe using pandas
df = pd.read_csv("ecommerce-session-bigquery.csv")

# show 5 first rows of the dataframe
df.head()
```

```python
# describe the data
df.describe()
```

```python
# looking for info (column name, data type, sum of rows and columns) from the data
df.info()
```

```python
# check missing value
df.isna().sum()
```

```python
# looking for column with missing value more than 50%
missing_values = df.isna().sum()
columns_with_missing_values = missing_values.index.tolist()

# Divide into two, that is categorical and numerical
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
numerical_columns = df.select_dtypes(exclude=['object']).columns.tolist()

# Filtering column that have missing value more than 50% from all of the data rows
categorical_columns_with_missing_values = [col for col in columns_with_missing_values if col in categorical_columns]
numerical_columns_with_missing_values = [col for col in columns_with_missing_values if col in numerical_columns]

print("Categorical Columns with Missing Values:")
print(categorical_columns_with_missing_values)

print("\nNumerical Columns with Missing Values:")
print(numerical_columns_with_missing_values)
```

```python
# From the numerical columns on data, i think "NaN" is not a missing value, but it must a 0, or nothing.
# I filled the NaN value with "0" as an integer
df[numerical_columns_with_missing_values] = df[numerical_columns_with_missing_values].fillna(0)
```

```python
# Check again missing value on categorical data
df[categorical_columns_with_missing_values].isna().sum()
```

```python
# Need to look for unique value on "currencyCode" and "pageTitle" column
print(df.currencyCode.unique())
print(df.pageTitle.unique())
```

```python
# Because of the "NaN" value is on the categorical data, I will fill it with mode() from each column
# Fill NaN values with mode for 'currencyCode' and 'pageTitle'
df['currencyCode'] = df['currencyCode'].fillna(df['currencyCode'].mode()[0])
df['pageTitle'] = df['pageTitle'].fillna(df['pageTitle'].mode()[0])
```

```python
# Looking for the dataframe missing value updated after cleansing
df.isna().sum()
```

## 1. Identify top products based on the total transaction revenue per day.

```python
# Do some feature engineering
# First, we need to convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'].astype(str), format='%Y%m%d')

# Create new columns 'year', 'month', and 'day'
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

# Print the updated DataFrame
df.head()
```

### Top 5 Products by Total Transaction Revenue per Day

```python
# Aggregate data by 'date' and 'productName' to get total transaction revenue
grouped_df_transPerDay = df.groupby(['date', 'v2ProductName']).agg({'totalTransactionRevenue': 'sum'}).reset_index()

# Identify top products based on total transaction revenue per day
top_products_per_day = grouped_df_transPerDay.groupby('date').apply(lambda x: x.nlargest(1, 'totalTransactionRevenue')).reset_index(drop=True)

# Sort the dataframe based on 'totalTransactionRevenue' in descending order and get the top 5
top_5_products_sorted = top_products_per_day.nlargest(5, 'totalTransactionRevenue')

# Print the top 5 products per day sorted by total transaction revenue
print(top_5_products_sorted)

plt.figure(figsize=(16, 8))

# Loop melalui setiap tanggal dan plot produk teratas untuk setiap tanggal
for date in top_5_products_sorted['date'].unique():
    data_for_date = top_5_products_sorted[top_5_products_sorted['date'] == date]
    plt.barh(data_for_date['v2ProductName'], data_for_date['totalTransactionRevenue'], label=str(date))

plt.xlabel('Total Transaction Revenue')
plt.ylabel('Product Name')
plt.title('Top 5 Products by Total Transaction Revenue per Day')
plt.legend(title='Date', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

Based on the above data, we have 5 products with the highest total transaction revenue in one day. In sequence, they are:

1. Youtube RFID Journal
2. Sport Bag
3. Collapsible Shoping Bag
4. Electronics Accessory Pouch
5. Google Pet Feeding Mat

### Top 5 Products by Average Transaction Revenue per Day

```python
# Aggregate data by 'date' and 'v2ProductName' to get total transaction revenue
grouped_df_avgTransPerDay = df.groupby(['date', 'v2ProductName']).agg({'totalTransactionRevenue': 'sum'}).reset_index()

# Calculate average total transaction revenue per day for each product
avg_revenue_per_day = grouped_df_avgTransPerDay.groupby('v2ProductName')['totalTransactionRevenue'].mean().reset_index()

# Sort the products based on average revenue per day in descending order
sorted_products = avg_revenue_per_day.sort_values(by='totalTransactionRevenue', ascending=False)

# Print the sorted products by average revenue per day
print(sorted_products)

plt.figure(figsize=(12, 6))

# Plot the top 10 products by average revenue per day
top_10_products = sorted_products.head(10)
plt.barh(top_10_products['v2ProductName'], top_10_products['totalTransactionRevenue'], color='skyblue')
plt.xlabel('Average Transaction Revenue per Day')
plt.ylabel('Product Name')
plt.title('Top 10 Products by Average Transaction Revenue per Day')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

Based on the above data, we have 5 products with the largest total transaction revenue per day. In sequence, they are:

1. YouTube RFID Journal
2. Google Tote Bag
3. Waterpoof Gear Bag
4. Google Women's Yoga Jacket Black
5. Collapsible Shopping Bag

## 2. Detect any anomalies, such as a sharp decrease or increase in the number of transactions for a specific product.

I use for specific products :
1. Youtube RFID Journal
2. Google Tote Bag
3. Waterproof Gear Bag
4. Google Women's Yoga Jacket Black
5. Collapsible Shopping Bag

```python
# set "date" column as an index
df.set_index('date', inplace=True)

# List of Specific Product
products_to_display = [
    'YouTube RFID Journal',
    'Google Tote Bag',
    'Waterproof Gear Bag',
    "Google Women's Yoga Jacket Black",
    'Collapsible Shopping Bag'
]

# Filtering dataframe based on list of specific product
filtered_df_timeSeriesProduct = df[df['v2ProductName'].isin(products_to_display)]

# Data Visualization by Month (Timeseries)
plt.figure(figsize=(16, 6))
filtered_df_timeSeriesProduct.groupby(['v2ProductName']).resample('d')['transactions'].sum().unstack('v2ProductName').plot()
plt.title('Time Series of Daily Transactions by Product Name')
plt.xlabel('Date')
plt.ylabel('Transactions')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.show()

# Data Visualization by Month (Timeseries)
plt.figure(figsize=(16, 6))
filtered_df_timeSeriesProduct.groupby(['v2ProductName']).resample('m')['transactions'].sum().unstack('v2ProductName').plot()
plt.title('Time Series of Monthly Transactions by Product Name')
plt.xlabel('Month')
plt.ylabel('Transactions')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.show()

# Data Visualization by Year (Timeseries)
plt.figure(figsize=(16, 6))
filtered_df_timeSeriesProduct.groupby(['v2ProductName']).resample('Y')['transactions'].sum().unstack('v2ProductName').plot()
plt.title('Time Series of Yearly Transactions by Product Name')
plt.xlabel('Year')
plt.ylabel('Transactions')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.show()
```

From the timeseries diagram, I obtained several insights:

1. In the Monthly Timeseries diagram, I observed that almost all products displayed experience a declining trend every month, except for the "Waterproof Gear Bag" product which experienced an increase in monthly transaction volume from August 2017 to early September 2017. However, it followed the same monthly trend, which is consistently decreasing, same with all products trend.

## 3. Identify the most profitable city or province based on the total transaction revenue.

```python
# Group by 'city' and calculate total transaction revenue
grouped_df_MostProfitCityRevenue = df.groupby('city')['totalTransactionRevenue'].sum().reset_index()

# Sort the dataframe based on 'totalTransactionRevenue' in descending order
sorted_df_MostProfitCityRevenue = grouped_df_MostProfitCityRevenue.sort_values(by='totalTransactionRevenue', ascending=False)

# Get the top 10 cities or provinces based on total transaction revenue
top_10_cities = sorted_df_MostProfitCityRevenue.head(11)

# Plotting
plt.figure(figsize=(12, 6))
plt.barh(top_10_cities['city'], top_10_cities['totalTransactionRevenue'], color='skyblue')
plt.xlabel('Total Transaction Revenue')
plt.ylabel('City')
plt.title('Top 10 Cities by Total Transaction Revenue')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest revenue at the top
plt.tight_layout()
plt.show()
```

From the above diagram, we can see that the top 10 cities generating the highest total transaction revenue in sequence are:

1. New York
2. San Francisco
3. Mountain View
4. Toronto
5. Chicago, etc.
